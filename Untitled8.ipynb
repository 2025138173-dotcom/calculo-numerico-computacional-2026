```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return x**3 - 9*x + 3

# Criando valores de x
x = np.linspace(-3, 3, 400)  # intervalo maior para visualizar melhor

# Calculando os valores de y
y = f(x)

# Plotando o gráfico
plt.figure()
plt.plot(x, y, label='f(x) = x³ - 9x + 3')

# Destacando o intervalo [0,1]
plt.axvline(0)
plt.axvline(1)

# Linha do eixo x
plt.axhline(0)

plt.title('Gráfico da função f(x) = x³ - 9x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

plt.show()
```

Etapa 01 – Onde parece estar localizada a raiz?
A raiz parece estar no intervalo entre 0 e 1. Isso acontece porque, nesse intervalo, a função muda de sinal. Pelo Teorema do Valor Intermediário, quando ocorre essa mudança de sinal, significa que existe pelo menos uma raiz dentro desse intervalo.

Existe mais de uma raiz?
Sim. Como se trata de uma função cúbica (com termo

```python
# Definindo a função
def f(x):
    return x**3 - 9*x + 3

# Definindo o intervalo
a = 0
b = 1

# Calculando os valores da função
fa = f(a)
fb = f(b)

# Mostrando os resultados
print("f(a) =", fa)
print("f(b) =", fb)
```

Etapa 02 – Existe mudança de sinal?
Sim, existe. O valor de f(a) é positivo, enquanto o valor de f(b) é negativo. Isso mostra que houve uma mudança de sinal entre esses dois pontos.

O intervalo escolhido é adequado?
Sim, o intervalo é adequado. Como ocorre essa mudança de sinal, isso indica que existe uma raiz dentro dele. Por isso, esse intervalo pode ser usado normalmente no método da bisseção para encontrar uma aproximação da raiz.

```python
# Definindo a função do método da bisseção
def bisseccao(f, a, b, n):

    if f(a) * f(b) >= 0:
        print("O método da bisseção não pode ser aplicado neste intervalo.")
        return None

    for i in range(n):
        m = (a + b) / 2  # ponto médio

        if f(a) * f(m) < 0:
            b = m
        else:
            a = m

    return m


# Definindo a função do exercício
def f(x):
    return x**3 - 9*x + 3


# Testando
print(bisseccao(f, 0, 1, 5))
print(bisseccao(f, 0, 1, 10))
print(bisseccao(f, 0, 1, 20))
```

```python
import pandas as pd

# Função do método da bisseção que cria uma tabela
def bisseccao_tabela(f, a, b, n):

    dados = []

    for i in range(1, n+1):
        c = (a + b) / 2
        fc = f(c)

        dados.append([i, a, b, c, fc])

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    tabela = pd.DataFrame(dados, columns=["Iteração", "a", "b", "c", "f(c)"])
    return tabela


# Função do exercício
def f(x):
    return x**3 - 9*x + 3


# Executar 10 iterações
tabela = bisseccao_tabela(f, 0, 1, 10)

tabela
```

Sim. A cada iteração o intervalo foi dividido pela metade, o que é justamente a característica principal do método da bisseção.

O resultado obtido parece coerente com o gráfico da função?
Sim. Observando o gráfico da função, é possível perceber que a curva cruza o eixo x entre 0 e 1, aproximadamente perto de 0,33. O valor encontrado pelo método da bisseção ficou bem próximo desse ponto, o que mostra que o resultado faz sentido e está de acordo com o gráfico.

Qual é a principal limitação desse método?
A principal limitação é que ele pode ser mais lento do que outros métodos numéricos, já que precisa de várias iterações para chegar a uma aproximação mais precisa.

Em quais situações ele pode ser mais adequado que outros métodos?
O método da bisseção é mais indicado quando queremos garantir que a raiz seja encontrada, quando sabemos apenas um intervalo onde há mudança de sinal, ou quando a função é mais complicada e difícil de derivar, tornando útil um método mais simples e seguro.
