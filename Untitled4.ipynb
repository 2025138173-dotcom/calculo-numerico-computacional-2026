```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - x - 2

x = np.linspace(-3, 3, 400)
y = f(x)

plt.axhline(0)
plt.plot(x, y)
plt.title("Gráfico da função no intervalo [-3, 3]")
plt.grid()
plt.show()
```

Parte 1 - Quantas raízes aparentes existem? Existe 1 raiz aparente.

Entre os números 1 e 2 da coordenada X, do lado positivo.

```python
for i in range(-3, 4):
    print(f"x = {i}, f(x) = {f(i)}")
```

```python
for i in range(-3, 3):
    if f(i) * f(i+1) < 0:
        print(f"Mudança de sinal entre {i} e {i+1}")
```

Parte 2 - Em qual intervalo ocorre mudança de sinal? No intervalo entre 1 e 2. Pois f(1)= -2 e f(2)= 4. 1 negativo e 2 positivo.

Existe mais de um intervalo válido? Não. Pois nos outros valores inteiros não ocorre mudança de sinal. Apenas no intervalo [-3, 3].

Qual intervalo você escolheria para estudar primeiro? Eu escolheria o intervalo que há mudança de sinal, [1, 2]. Pelo Teorema do Valor Intermediário, isso garante que existe pelo menos uma raiz dentro desse intervalo. E é o único intervalo válido encontrado.



```python
a = 1
b = 2

for _ in range(5):  # 5 iterações
    m = (a + b) / 2
    print(f"Intervalo atual: [{a}, {b}]")
    print(f"Ponto médio: {m}")

    if f(a) * f(m) < 0:
        b = m
    else:
        a = m
```

Parte 03 - O intervalo pode ser reduzido? Sim. Podemos diminuir o intervalo pela metade, tornando a busca mais precisa.

A raiz parece estar mais próxima de qual extremo? A raiz está mais próxima de 1,5 do que de 2, e dentro do intervalo inicial, ela está mais próxima do extremo esquerdo (1) do que do 2.

Que tipo de método estamos construindo intuitivamente? Método da Bisseção, já que começamos com um intervalo onde há mudança de sinal; Calculamos o ponto médio; Verificamos em qual metade ocorre mudança de sinal; Repetimos o processo. Ou seja, dividimos o intervalo ao meio sucessivamente até encontrar uma boa aproximação da raiz.

Parte 04 - Por que a mudança de sinal indica existência de raiz? Porque, se a função é contínua, quando ela passa de um valor negativo para positivo (ou vice-versa), ela necessariamente precisa cruzar o eixo x em algum ponto. Isso é garantido pelo: Teorema do Valor Intermediário.

   A ausência de mudança de sinal garante ausência de raiz?
 Não necessariamente. Por exemplo, existe uma raiz onde a função apenas "toca" o eixo x e volta (raiz de multiplicidade par), sem trocar o sinal. Exemplo: f(𝑥)=x**2. Ela tem raiz em x=0, mas não há mudança de sinal, porque a função é positiva dos dois lados.

Por que é importante localizar antes de aplicar um método iterativo? Métodos iterativos precisam de um bom ponto inicial ou intervalo adequado. Se escolhermos um intervalo errado, o método pode não convergir, demorar muito encontrar outra raiz diferente. No método da bisseção, por exemplo, é obrigatório que exista mudança de sinal no intervalo inicial.

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return (x - 1)**2

# Valores para o gráfico
x = np.linspace(-1, 3, 400)
y = g(x)

plt.axhline(0)
plt.plot(x, y)
plt.title("Gráfico de g(x) = (x - 1)^2")
plt.grid()
plt.show()

# Testando sinais
for i in range(-1, 3):
    print(f"x = {i}, g(x) = {g(i)}")
```

Parte 5 - Desafio extra, Existe raiz? Sim. Quando X=1.

Há mudança de sinal? Não. A função é sempre maior ou igual a zero, pois é um quadrado. Ela apenas toca o eixo x em x = 1, mas não atravessa. E percebe-se que os valores de g(x) estão todos positivos.Então não há mudança de sinal.

Um método baseado apenas em sinal funcionaria? Não. Métodos como o da bisseção precisam de mudança de sinal no intervalo inicial. Como não há troca de sinal, esse tipo de método não detectaria essa raiz.
