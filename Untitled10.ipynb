```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return x**3 - x - 2

# Criando valores de x no intervalo [1, 2]
x = np.linspace(1, 2, 100)
y = f(x)

# Plotando o gráfico
plt.figure()
plt.plot(x, y)
plt.axhline(0)  # eixo x
plt.axvline(0)  # eixo y
plt.title('Gráfico de f(x) = x³ - x - 2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

plt.show()
```

Visualização da função – Observando o gráfico, a raiz parece estar situada próxima de x ≈ 0,74, onde a função cruza o eixo y = 0.

Número de raízes no intervalo – No intervalo analisado, a função possui apenas uma raiz, não há outros pontos em que ela cruza o eixo x.

```python
import pandas as pd

# Definindo a função
def f(x):
    return x**3 - x - 2

# Intervalo inicial
a = 1
b = 2

# Número de iterações
n = 10

# Lista para armazenar os dados
dados = []

for i in range(1, n + 1):
    c = (a + b) / 2
    fc = f(c)

    # Guardando os valores
    dados.append([i, a, b, c, fc])

    # Método da bissecção
    if f(a) * fc < 0:
        b = c
    else:
        a = c

# Criando tabela
tabela = pd.DataFrame(dados, columns=["Iteração", "a", "b", "c", "f(c)"])

# Exibindo
print(tabela)
```

```python
import pandas as pd

# Função
def f(x):
    return x**3 - x - 2

# Derivada
def df(x):
    return 3*x**2 - 1

# Valor inicial (próximo da raiz da bissecção)
x = 1.52

# Número de iterações
n = 10

# Lista para armazenar resultados
dados = []

for i in range(1, n + 1):
    fx = f(x)
    dfx = df(x)

    # Método de Newton
    x_new = x - fx / dfx

    # Erro absoluto
    erro = abs(x_new - x)

    # Armazenar dados
    dados.append([i, x, fx, erro])

    # Atualizar x
    x = x_new

# Criar tabela
tabela = pd.DataFrame(dados, columns=["Iteração", "x_n", "f(x_n)", "erro"])

# Exibir tabela
tabela
```

```python

import pandas as pd

# Função e derivada
def f(x):
    return x**3 - x - 2

def df(x):
    return 3*x**2 - 1

# -------------------------
# MÉTODO DA BISSECÇÃO
# -------------------------
a = 1
b = 2
tol = 1e-6
max_iter = 100

iter_bis = 0

while (b - a)/2 > tol:
    c = (a + b)/2
    if f(a)*f(c) < 0:
        b = c
    else:
        a = c
    iter_bis += 1

raiz_bis = (a + b)/2
erro_bis = abs(f(raiz_bis))

# -------------------------
# MÉTODO DE NEWTON
# -------------------------
x = 1.52  # chute inicial
iter_newton = 0

for i in range(max_iter):
    x_new = x - f(x)/df(x)

    if abs(x_new - x) < tol:
        break

    x = x_new
    iter_newton += 1

raiz_newton = x
erro_newton = abs(f(x))

# -------------------------
# COMPARAÇÃO
# -------------------------
comparacao = pd.DataFrame({
    "Método": ["Bissecção", "Newton"],
    "Raiz aproximada": [raiz_bis, raiz_newton],
    "Iterações": [iter_bis, iter_newton],
    "Erro final": [erro_bis, erro_newton]
})

comparacao
```

```python
import pandas as pd

# Função e derivada
def f(x):
    return x**3 - x - 2

def df(x):
    return 3*x**2 - 1

# Lista de chutes iniciais
chutes = [1, 2, 20]

# Parâmetros
tol = 1e-6
max_iter = 50

# Lista para resultados
resultados = []

for x0 in chutes:
    x = x0
    iteracoes = 0

    for i in range(max_iter):
        # Evitar divisão por zero
        if df(x) == 0:
            break

        x_new = x - f(x)/df(x)

        if abs(x_new - x) < tol:
            x = x_new
            break

        x = x_new
        iteracoes += 1

    erro = abs(f(x))

    resultados.append([x0, x, iteracoes, erro])

# Criar tabela
tabela = pd.DataFrame(resultados, columns=[
    "Chute inicial (x0)", "Raiz aproximada", "Iterações", "Erro final"
])

# Exibir
tabela
```

Experimento extra – O método de Newton não garante convergência em todos os casos; ele pode falhar dependendo da função e do valor inicial escolhido.

Influência do chute inicial – O desempenho do método depende bastante do valor inicial. Se o ponto de partida estiver próximo da raiz, o método tende a encontrar a solução rapidamente; caso contrário, pode demorar ou não convergir.

Reflexão sobre a velocidade – Newton geralmente chega à raiz mais rapidamente porque utiliza a derivada da função para ajustar cada aproximação, enquanto a bissecção apenas reduz o intervalo pela metade a cada passo.

Quando a bissecção é vantajosa – Esse método é útil quando se quer garantir que existe uma raiz dentro de um intervalo ou quando a derivada da função não está disponível, já que não depende dela.

Escolha em problemas de engenharia – Na prática, costuma-se combinar os dois métodos: a bissecção é usada primeiro para localizar um bom ponto inicial e, em seguida, Newton é aplicado para chegar à solução de forma rápida e eficiente.
