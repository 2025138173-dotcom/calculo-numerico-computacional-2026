```python
# Função e derivada
def f(x):
    return x**3 - 9*x + 3

def df(x):
    return 3*x**2 - 9


# ---------------------------
# Bissecção
# ---------------------------
def bisseccao(a, b, tol=1e-6):
    iteracoes = 0

    while abs(b - a) > tol:
        c = (a + b) / 2

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

        iteracoes += 1

    return (a + b)/2, iteracoes


# ---------------------------
# Newton
# ---------------------------
def newton(x0, tol=1e-6):
    iteracoes = 0

    while True:
        x1 = x0 - f(x0)/df(x0)

        if abs(x1 - x0) < tol:
            return x1, iteracoes

        x0 = x1
        iteracoes += 1


# ---------------------------
# Secante
# ---------------------------
def secante(x0, x1, tol=1e-6):
    iteracoes = 0

    while True:
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))

        if abs(x2 - x1) < tol:
            return x2, iteracoes

        x0, x1 = x1, x2
        iteracoes += 1


# ---------------------------
# Execução
# ---------------------------

# Escolhendo valores iniciais próximos da raiz ~1.52
raiz_bis, it_bis = bisseccao(1, 2)
raiz_new, it_new = newton(1.5)
raiz_sec, it_sec = secante(1, 2)

# ---------------------------
# Tabela final
# ---------------------------
print("\nComparação de Métodos\n")
print("Método\t\tIterações até convergir")
print(f"Bissecção\t{it_bis}")
print(f"Newton\t\t{it_new}")
print(f"Secante\t\t{it_sec}")

# Mostrar raízes (opcional)
print("\nRaízes aproximadas:")
print(f"Bissecção: {raiz_bis:.6f}")
print(f"Newton:    {raiz_new:.6f}")
print(f"Secante:   {raiz_sec:.6f}")
```

Qual método convergiu mais rapidamente?

O método que convergiu mais rapidamente foi o método de Newton. Isso ocorre porque ele utiliza a derivada da função, permitindo uma aproximação mais eficiente da raiz a cada iteração. Em geral, ele precisa de poucas iterações (cerca de 4 a 6) para atingir boa precisão. O método da secante também é rápido, mas ligeiramente menos eficiente, enquanto o método da bissecção é o mais lento, embora seja o mais seguro.

A secante sempre converge? Teste com diferentes valores iniciais.

Não, o método da secante não garante convergência em todos os casos. Seu desempenho depende dos valores iniciais escolhidos. Quando os valores estão próximos da raiz, ele converge rapidamente. Porém, se estiverem distantes ou mal escolhidos, o método pode oscilar ou até divergir.

Em quais situações o método da secante pode falhar?

O método da secante pode falhar quando os valores da função em duas iterações consecutivas são muito próximos, causando divisão por um número muito pequeno e gerando instabilidade numérica. Também pode falhar quando os valores iniciais estão longe da raiz ou quando a função apresenta comportamento irregular, dificultando a convergência.

Compare a necessidade de informação adicional entre os métodos.

O método da bissecção requer apenas um intervalo inicial com mudança de sinal da função, garantindo a existência de uma raiz. O método de Newton exige um valor inicial e o cálculo da derivada da função. Já o método da secante precisa de dois valores iniciais, mas não requer o uso da derivada, sendo uma alternativa intermediária.

Se você estivesse desenvolvendo um software de engenharia para encontrar raízes de funções desconhecidas, qual método escolheria e por quê?

Em um software de engenharia, o ideal é utilizar uma combinação de métodos. Primeiro, usar o método da bissecção para garantir a convergência e identificar um intervalo com raiz. Depois, aplicar o método de Newton ou da secante para acelerar a obtenção da solução. Essa abordagem combina segurança e eficiência, tornando o processo mais robusto.
