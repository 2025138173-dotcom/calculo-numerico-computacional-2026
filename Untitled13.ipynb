```python
def gauss_elimination(A, b):
    n = len(A)

    # Etapa 1: Eliminação (forma triangular)
    for i in range(n):
        # Pivô (evitar divisão por zero)
        if A[i][i] == 0:
            raise ValueError("Pivô zero encontrado!")

        for j in range(i + 1, n):
            fator = A[j][i] / A[i][i]

            for k in range(n):
                A[j][k] -= fator * A[i][k]

            b[j] -= fator * b[i]

    # Etapa 2: Substituição regressiva
    x = [0] * n

    for i in range(n - 1, -1, -1):
        soma = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x[i] = (b[i] - soma) / A[i][i]

    return x
```

```python
import numpy as np

def gauss_elimination(A, b):
    n = len(A)

    # Converter para float (evita erro de divisão inteira)
    A = [list(map(float, row)) for row in A]
    b = list(map(float, b))

    # Etapa 1: Eliminação
    for i in range(n):
        if A[i][i] == 0:
            raise ValueError("Pivô zero encontrado!")

        for j in range(i + 1, n):
            fator = A[j][i] / A[i][i]

            for k in range(n):
                A[j][k] -= fator * A[i][k]

            b[j] -= fator * b[i]

    # Etapa 2: Substituição regressiva
    x = [0] * n

    for i in range(n - 1, -1, -1):
        soma = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x[i] = (b[i] - soma) / A[i][i]

    return x


# Sistema
A = [
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
]

b = [8, -11, -3]

# Resultado com Gauss
sol_gauss = gauss_elimination(A, b)

# Resultado com NumPy
A_np = np.array(A, dtype=float)
b_np = np.array(b, dtype=float)

sol_numpy = np.linalg.solve(A_np, b_np)

# Comparação
print("Solução (Gauss):", sol_gauss)
print("Solução (NumPy):", sol_numpy)
```

O método de Gauss funciona corretamente quando o sistema possui solução única, ou seja, quando o número de equações é igual ao número de incógnitas, os pivôs da matriz não são zero e o sistema é possível e determinado.

```python
def gauss_elimination_detect(A, b):
    n = len(A)

    A = [list(map(float, row)) for row in A]
    b = list(map(float, b))

    # Eliminação
    for i in range(n):
        # Evitar pivô zero simples (sem pivotamento)
        if A[i][i] == 0:
            continue

        for j in range(i + 1, n):
            fator = A[j][i] / A[i][i]

            for k in range(n):
                A[j][k] -= fator * A[i][k]

            b[j] -= fator * b[i]

    # Verificação do tipo de sistema
    for i in range(n):
        if all(abs(A[i][j]) < 1e-9 for j in range(n)):
            if abs(b[i]) < 1e-9:
                return "Infinitas soluções"
            else:
                return "Sistema impossível"

    # Substituição regressiva (se tiver solução única)
    x = [0] * n
    for i in range(n - 1, -1, -1):
        soma = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x[i] = (b[i] - soma) / A[i][i]

    return x


# 🔬 Sistema do teste
A = [
    [1, 1],
    [2, 2]
]

b = [2, 4]

resultado = gauss_elimination_detect(A, b)

print("Resultado:", resultado)
```
