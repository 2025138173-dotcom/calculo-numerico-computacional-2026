```python
def f(x):
    return x**3 - x - 2


def bisseccao(f, a, b, n):
    for i in range(n):
        c = (a + b) / 2
        print("Iteração:", i, "C =", c)

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return (a + b) / 2
    # Chamando a função
raiz = bisseccao(f, 1, 2, 10)

print("A raiz aproximada é:", raiz)
```

O que aconteceu?

Quando executamos o primeiro código até a linha return (a + b) / 2, não aparece nenhum resultado na tela. Isso acontece porque a função apenas foi criada, mas ainda não foi chamada para rodar. Ou seja, o programa sabe o que fazer, mas ainda não recebeu o comando para executar a função.

Quando usamos o print, a função passa a ser executada e o resultado é mostrado na tela, exibindo o valor aproximado da raiz da equação.

```python
def f(x):
    return x**3 - x - 2


def bisseccao(f, a, b, n):

    print("Iter |   a   |   b   |   c   |  f(c)")
    print("--------------------------------------")

    for i in range(n):

        c = (a + b) / 2

        print(i, "|", round(a,6), "|", round(b,6), "|", round(c,6), "|", round(f(c),6))

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return (a + b) / 2


raiz = bisseccao(f, 1, 2, 10)

print("\nRaiz aproximada:", raiz)
bisseccao(f,1,2,5)

bisseccao(f,1,2,10)

bisseccao(f,1,2,20)
```

O que acontece com o resultado?

Quando aumentamos o número de iterações no método da bisseção, o intervalo é dividido mais vezes. Isso faz com que o valor encontrado fique cada vez mais próximo da raiz real da equação. Por isso, ao usar 20 iterações, o resultado acaba sendo mais preciso do que quando usamos apenas 5 ou 10.
